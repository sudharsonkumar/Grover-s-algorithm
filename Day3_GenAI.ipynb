```python
import numpy as np
# Vectors
v1= np.array([1, 2, 3])
v2= np.array([4, 5, 6])
#Vector addition and subtraction
v_add=v1 + v2
v_sub = v1 - v2
print("Vector addition:", v_add)
print("Vector subtraction:", v_sub)
```

```python
#Scalar multiplication
v_scalar = 3 * v1
print("Scalar multiplication:", v_scalar)
```

```python
#Matrices
m1=np.array([[1, 2], [3, 4]])
m2=np.array([[5, 6], [7, 8]])
#Matrix addition
m_add=m1+m2
print("Matrix addition:", m_add)
#Matrix subtraction
m_sub=m1-m2
print("Matrix subtraction:", m_sub)
#Scalar multiplication
m_scalar=3*m1
print("Scalar multiplication:", m_scalar)
#Matrix multiplication
m_mul=np.dot(m1, m2)
print("Matrix multiplication:", m_mul)
#Matrix transpose
m_trans=m1.T
print("Matrix transpose:", m_trans)
```

```python
import matplotlib.pyplot as plt

#Define the quadratic function and its gradient
def f(x):
    return x**2

def grad_f(x):
    return 2*x

#Gradient descent parameters
x = 10
learning_rate = 0.1
iterations = 20

#Lists to stor the values for plotting
x_values = [x]
f_values = [f(x)]

#Gradient descent loop
for i in range(iterations):
    grad = grad_f(x)
    x = x - learning_rate * grad
    x_values.append(x)
    f_values.append(f(x))

#Plot the function and the optimization path
x_range=np.linspace(-10, 10, 400)
plt.plot(x_range, f(x_range), label='$f(x) = x^2$')
plt.scatter(x_values, f_values, color='red')
plt.plot(x_values, f_values, color='red', label='Gradient Descent Path')
plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.title('Gradient Descent Optimization')
plt.legend()
plt.grid(True)
plt.show()
```
